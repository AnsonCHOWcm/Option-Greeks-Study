```python
import mibian
from matplotlib import pyplot as plt
import pandas as pd
```

```python
time_point_ls = []
gamma_ls = []
vega_ls = []

for i in range(0,360,5):
    option_price = mibian.BS([100,100,5,360 - i], volatility = 10)
    gamma_ls.append(option_price.gamma)
    vega_ls.append(option_price.vega)
    time_point_ls.append(360-i)
    
gamma_df = pd.DataFrame()
gamma_df['gamma'] = gamma_ls
gamma_df.index = time_point_ls

vega_df = pd.DataFrame()
vega_df['vega'] = vega_ls
vega_df.index = time_point_ls
```

```python
gamma_df.plot()
```

```python
vega_df.plot()
```
